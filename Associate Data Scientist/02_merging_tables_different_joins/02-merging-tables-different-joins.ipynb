```python
import pandas as pd

# Contoh data
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Nama': ['Ari', 'Budi', 'Citra']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Skor': [85, 90, 78]
})

# Menampilkan kedua tabel
print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)
```

```python
df_inner = df1.merge(df2, on='ID', how='inner')
print("Inner Join Result:")
print(df_inner)
```

```python
df_left = df1.merge(df2, on='ID', how='left')
print("Left Join Result:")
print(df_left)
```

```python
df_right = df1.merge(df2, on='ID', how='right')
print("Right Join Result:")
print(df_right)
```

```python
df_outer = df1.merge(df2, on='ID', how='outer')
print("Outer Join Result:")
print(df_outer)
```
